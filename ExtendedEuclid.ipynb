Der Euklidische Algorithmus in seiner einfachsten Form ermittelt den Groessten Gemeinsamen Teiler (ggt) g von zwei  Zahlen A und B:

g teilt A <==> es existiert c: g*c==A

Man schreibt g|A 

Ein einfacher Satz:

g|A und g|B und A>B, dann g|(A-B).
Da dies fuer alle gemeinsamen Teiler von A und B gilt, gilt es auch fuer den groessten gemeinsamen Teiler.
Wenn A==B oder B==0 dann ist der ggt von A und B gleich A.
